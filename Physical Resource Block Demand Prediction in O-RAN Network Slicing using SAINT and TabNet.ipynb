# PRB Demand Prediction in O-RAN Network Slicing using SAINT and TabNet

## Overview

Efficient allocation of Physical Resource Blocks (PRBs) is critical in 
Open Radio Access Networks (O-RAN) to ensure Quality of Service (QoS) 
while avoiding resource over-provisioning.

This notebook implements transformer-based and tabular deep learning 
models to predict requested PRBs at the slice level using telemetry 
data derived from the ColO-RAN experimental framework.

Models implemented:
- SAINT (Self-Attention and Intersample Transforine comparisons

Evaluation metrics:
- Mean Squared Error (MSE)
- R² Score
- Mean Absolute Percentage Error (MAPE)

## Research Context

This work is based on the ColO-RAN dataset:

M. Polese et al., "ColO-RAN: Developing machine learning-based xApps 
for open RAN closed-loop control on programmable experimental platforms,"
IEEE Transactions on Mobile Computing, 2022.

The dataset represents a multi-cell 5G scenario with:
- 7 base stations
- 42 user equipments
- Traffic classes: eMBB, URLLC, and mMTC

For this study, 78,561 slice-level telemetry records were aggregated
from 41 CSV logs located in `shed0/tr0/exp1/`.

The data has been processed into a structured tabular dataset
(`slices.csv`) for supervised PRB demand prediction.

## Problem Formulation

Given slice-level telemetry features:

X = {traffic statistics, QoS indicators, RAN metrics, historical utilization}

Predict:

y = Requested Physical Resource Blocks (PRBs)

This is formulated as a supervised regression problem.

Objective:
Minimize prediction error while maintaining high generalization performance.

## Data Preprocessing

The dataset undergoes the following preprocessing steps:

1. Handling missing values
2. Feature normalization / standardization
3. Train-test split (80/20)
4. Conversion to tensor format (for SAINT)

Preprocessing ensures stable training and improved model convergence.

## Model Training

Both models are trained using:

- Optimizer: Adam
- Loss Function: Mean Squared Error (MSE)
- Batch training for stability

Training performance is monitored using validation metrics.

## Model Evaluation

Performance is evaluated using:

- Mean Squared Error (MSE)
- R² Score
- Mean Absolute Percentage Error (MAPE)

These metrics measure:

- Absolute prediction accuracy
- Variance explanation capability
- Relative percentage error

Comparison between SAINT and TabNet provides insight into
transformer effectiveness for O-RAN slicing scenarios.

## Results and Discussion

The experimental results demonstrate the capability of 
transformer-based tabular models in learning complex slice-level 
resource demand patterns.

Key observations:
- SAINT captures feature interactions effectively
- TabNet provides competitive performance with interpretability
- Transformer-based approaches are promising for near-RT RIC xApp deployment

These findings support the feasibility of intelligent PRB 
prediction in O-RAN slicing environments.

## SAINT Model Architecture

SAINT (Self-Attention and Intersample Attention Transformer) is a transformer-based 
architecture specifically designed for tabular data.

Key characteristics:
- Feature-wise attention (column attention)
- Row-wise attention (intersample relationships)
- Improved representation learning for heterogeneous features

SAINT enables modeling complex dependencies between slice-level metrics,
making it suitable for PRB demand prediction.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score
import time

# Custom Dataset
class TabularDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)
    
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

# Define SAINT model
class SAINT(nn.Module):
    def __init__(self, num_features, d_model=32, n_heads=4, n_layers=2, dropout=0.1):
        super().__init__()
        self.num_features = num_features
        self.d_model = d_model
        self.embedding = nn.Linear(1, d_model)
        self.pos_emb = nn.Parameter(torch.randn(1, num_features, d_model))
        encoder_layer = nn.TransformerEncoderLayer(d_model, n_heads, dim_feedforward=4*d_model, dropout=dropout)
        self.row_encoder = nn.TransformerEncoder(encoder_layer, num_layers=n_layers)
        self.intersample_attn = nn.MultiheadAttention(d_model, n_heads, dropout=dropout)
        self.mlp = nn.Sequential(
            nn.LayerNorm(d_model),
            nn.Linear(d_model, d_model//2),
            nn.ReLU(),
            nn.Linear(d_model//2, 1)
        )

    def forward(self, x):
        B, N = x.size()
        x = x.unsqueeze(-1)
        x = self.embedding(x)
        x = x + self.pos_emb
        x = x.transpose(0, 1)
        x = self.row_encoder(x)
        x = x.transpose(0, 1)
        x_t = x.transpose(0, 1)
        x2, _ = self.intersample_attn(x_t, x_t, x_t)
        x2 = x2.transpose(0, 1)
        agg = x2.mean(dim=1)
        return self.mlp(agg).squeeze(-1)

# --- 1. Load & preprocess ---
df = pd.read_csv('slices.csv')

# Filter sum_requested_prbs > 0
df = df[df['sum_requested_prbs'] > 0]

# Log-transform target
df['sum_requested_prbs'] = np.log1p(df['sum_requested_prbs'])

# Feature engineering (timestamp cyclic features and raw timestamp)
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['hour_sin'] = np.sin(2 * np.pi * df['Timestamp'].dt.hour / 24)
df['hour_cos'] = np.cos(2 * np.pi * df['Timestamp'].dt.hour / 24)
df['day_sin'] = np.sin(2 * np.pi * df['Timestamp'].dt.dayofweek / 7)
df['day_cos'] = np.cos(2 * np.pi * df['Timestamp'].dt.dayofweek / 7)
df['timestamp_numeric'] = df['Timestamp'].astype('int64') // 10**9  # Unix timestamp in seconds

# Drop irrelevant columns, including sum_granted_prbs
df.drop(columns=['Timestamp', 'IMSI', 'RNTI', 'sum_granted_prbs'], inplace=True, errors='ignore')

# Define features
features = [c for c in df.columns if c != 'sum_requested_prbs']
print("Features used:", features)

# Select features and target
X = df[features].fillna(df[features].median()).values
y = df['sum_requested_prbs'].values

# Scale features
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Initialize variables
n_splits = 5
kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
mse_scores = []
r2_scores = []
mape_scores_log = []
mape_scores_orig = []
all_y_true = []
all_y_pred = []
all_train_losses = []
all_val_losses = []
fold_times = []

# --- 2. K-fold cross-validation ---
for fold, (train_idx, val_idx) in enumerate(kf.split(X), 1):
    print(f"\nFold {fold}/{n_splits}")
    X_tr, X_val = X[train_idx], X[val_idx]
    y_tr, y_val = y[train_idx], y[val_idx]

    # Dataset + DataLoader
    train_loader = DataLoader(TabularDataset(X_tr, y_tr), batch_size=64, shuffle=True)
    val_loader = DataLoader(TabularDataset(X_val, y_val), batch_size=128)

    # Initialize model and optimizer
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = SAINT(num_features=X_tr.shape[1]).to(device)
    opt = optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-5)
    crit = nn.MSELoss()

    # Define training and evaluation functions
    def train_epoch():
        model.train()
        total_loss = 0
        for xb, yb in train_loader:
            xb, yb = xb.to(device), yb.to(device)
            opt.zero_grad()
            loss = crit(model(xb), yb)
            loss.backward()
            opt.step()
            total_loss += loss.item() * len(xb)
        return total_loss / len(train_loader.dataset)

    @torch.no_grad()
    def eval_epoch():
        model.eval()
        mse = 0
        for xb, yb in val_loader:
            xb, yb = xb.to(device), yb.to(device)
            mse += crit(model(xb), yb).item() * len(xb)
        return mse / len(val_loader.dataset)

    # Train for 30 epochs
    train_losses = []
    val_losses = []
    best_val_loss = float('inf')
    start_time = time.time()

    for epoch in range(1, 31):
        tr_loss = train_epoch()
        val_loss = eval_epoch()
        train_losses.append(tr_loss)
        val_losses.append(val_loss)
        if val_loss < best_val_loss:
            best_val_loss = val_loss
        if epoch % 10 == 0 or epoch == 1:
            print(f"Epoch {epoch:02d} — Train MSE: {tr_loss:.4f}, Val MSE: {val_loss:.4f}")

    end_time = time.time()
    fold_times.append(end_time - start_time)
    all_train_losses.append(train_losses)
    all_val_losses.append(val_losses)

    # Save training curve for this fold
    plt.figure(figsize=(10, 5))
    plt.plot(train_losses, label='Train MSE')
    plt.plot(val_losses, label='Validation MSE')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title(f'Training vs Validation MSE (Fold {fold})')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f'saint_training_curve_fold{fold}.png', dpi=300)
    plt.close()

    # Calculate metrics for this fold
    @torch.no_grad()
    def calculate_metrics():
        model.eval()
        y_true = []
        y_pred = []
        for xb, yb in val_loader:
            xb, yb = xb.to(device), yb.to(device)
            pred = model(xb)
            y_true.extend(yb.cpu().numpy())
            y_pred.extend(pred.cpu().numpy())
        y_true = np.array(y_true)
        y_pred = np.array(y_pred)
        mse = mean_squared_error(y_true, y_pred)
        r2 = r2_score(y_true, y_pred)
        # MAPE on log scale
        mape_log = np.mean(np.abs((y_true - y_pred) / (y_true + 1e-6))) * 100
        # MAPE on original scale
        y_true_orig = np.expm1(y_true)
        y_pred_orig = np.expm1(y_pred)
        mape_orig = np.mean(np.abs((y_true_orig - y_pred_orig) / (y_true_orig + 1e-6))) * 100
        return mse, r2, mape_log, mape_orig, y_true, y_pred

    mse_fold, r2_fold, mape_log_fold, mape_orig_fold, y_true_fold, y_pred_fold = calculate_metrics()
    mse_scores.append(mse_fold)
    r2_scores.append(r2_fold)
    mape_scores_log.append(mape_log_fold)
    mape_scores_orig.append(mape_orig_fold)
    all_y_true.extend(y_true_fold)
    all_y_pred.extend(y_pred_fold)
    print(f"Fold {fold} — Validation MSE: {mse_fold:.4f}, R²: {r2_fold:.4f}, MAPE (log): {mape_log_fold:.2f}%, MAPE (original): {mape_orig_fold:.2f}%")

    # Per-fold actual vs. predicted plot (original scale)
    plt.figure(figsize=(12, 6))
    plt.plot(np.expm1(y_true_fold), label='Actual', color='blue')
    plt.plot(np.expm1(y_pred_fold), label='Predicted', color='orange')
    plt.title(f'Fold {fold}: Actual vs. Predicted sum_requested_prbs (Original Scale)')
    plt.xlabel('Validation Samples')
    plt.ylabel('sum_requested_prbs')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f'saint_actual_vs_predicted_fold{fold}.png', dpi=300)
    plt.show()

# --- 3. Summarize cross-validated metrics ---
mean_mse = np.mean(mse_scores)
std_mse = np.std(mse_scores)
mean_r2 = np.mean(r2_scores)
std_r2 = np.std(r2_scores)
mean_mape_log = np.mean(mape_scores_log)
std_mape_log = np.std(mape_scores_log)
mean_mape_orig = np.mean(mape_scores_orig)
std_mape_orig = np.std(mape_scores_orig)
print(f"\nCross-validated Metrics (5-fold):")
print(f"Mean MSE: {mean_mse:.4f} ± {std_mse:.4f}")
print(f"Mean R²: {mean_r2:.4f} ± {std_r2:.4f}")
print(f"Mean MAPE (log scale): {mean_mape_log:.2f}% ± {std_mape_log:.2f}%")
print(f"Mean MAPE (original scale): {mean_mape_orig:.2f}% ± {std_mape_orig:.2f}%")

# Combined actual vs. predicted plot (original scale)
all_y_true = np.array(all_y_true)
all_y_pred = np.array(all_y_pred)
plt.figure(figsize=(12, 6))
plt.plot(np.expm1(all_y_true), label='Actual', color='blue')
plt.plot(np.expm1(all_y_pred), label='Predicted', color='orange')
plt.title('Combined: Actual vs. Predicted sum_requested_prbs (Original Scale)')
plt.xlabel('Validation Samples (Across All Folds)')
plt.ylabel('sum_requested_prbs')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('saint_combined_actual_vs_predicted.png', dpi=300)
plt.show()

# Combined residual plot (original scale)
residuals = np.expm1(all_y_true) - np.expm1(all_y_pred)
plt.figure(figsize=(12, 6))
plt.scatter(np.expm1(all_y_true), residuals, alpha=0.5)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Actual sum_requested_prbs')
plt.ylabel('Residuals')
plt.title('Combined Residual Plot (Original Scale)')
plt.grid(True)
plt.tight_layout()
plt.savefig('saint_combined_residual.png', dpi=300)
plt.show()

# Plot training time per fold
plt.figure(figsize=(8, 4))
plt.bar(range(1, n_splits + 1), fold_times, color='skyblue')
plt.xlabel('Fold')
plt.ylabel('Training Time (seconds)')
plt.title('Training Time per Fold')
plt.tight_layout()
plt.savefig('saint_training_times.png', dpi=300)
plt.show()
```

## TabNet Model

TabNet is a deep learning architecture for tabular data that uses:

- Sequential attention mechanism
- Feature selection at each decision step
- Sparse feature masking

It provides interpretability and efficient feature utilization.

```python
import pandas as pd
import numpy as np
from pytorch_tabnet.tab_model import TabNetRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import TimeSeriesSplit
import matplotlib.pyplot as plt
import warnings
import time

warnings.filterwarnings('ignore')

# Load Data
df = pd.read_csv('slices.csv')
print("Initial dataset shape:", df.shape)
print("Number of zeros in sum_requested_prbs:", (df['sum_requested_prbs'] == 0).sum())
print("Number of NaN in sum_requested_prbs:", df['sum_requested_prbs'].isna().sum())
print("Number of negative values:", (df['sum_requested_prbs'] < 0).sum())
print("Number of duplicates:", df.duplicated().sum())
print("Available columns:", df.columns.tolist())

# Filter non-zero sum_requested_prbs and log transform
df = df[df['sum_requested_prbs'] > 0]
df['sum_requested_prbs'] = np.log1p(df['sum_requested_prbs'])
print("After filtering zeros, dataset shape:", df.shape)
print("sum_requested_prbs distribution (filtered):\n", df['sum_requested_prbs'].describe())

# Convert Timestamp to datetime and extract cyclic features
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['hour_sin'] = np.sin(2 * np.pi * df['Timestamp'].dt.hour / 24)
df['hour_cos'] = np.cos(2 * np.pi * df['Timestamp'].dt.hour / 24)
df['day_sin'] = np.sin(2 * np.pi * df['Timestamp'].dt.dayofweek / 7)
df['day_cos'] = np.cos(2 * np.pi * df['Timestamp'].dt.dayofweek / 7)

# Drop irrelevant columns
drop_columns = ['IMSI', 'RNTI', 'sum_granted_prbs']
df.drop(columns=drop_columns, inplace=True, errors='ignore')

# Define features and target
features = [col for col in df.columns if col != 'sum_requested_prbs' and col != 'Timestamp']
target = 'sum_requested_prbs'
print("Selected features:", features)

# Handle NaN values
print("NaN count in df before handling:\n", df[features].isna().sum())
df[features + [target]] = df[features + [target]].fillna(df[features + [target]].median())
print("After handling NaN, dataset shape:", df.shape)

# Scale numerical features
scaler = StandardScaler()
df[features] = scaler.fit_transform(df[features])

# Time-based split
tscv = TimeSeriesSplit(n_splits=5)
r2_scores = []
mse_scores = []
mape_scores_log = []
mape_scores_orig = []

for fold, (train_index, test_index) in enumerate(tscv.split(df), 1):
    print(f"\nFold {fold}/5")
    train_df = df.iloc[train_index]
    test_df = df.iloc[test_index]

    X_train = train_df[features].values
    y_train = train_df[target].values.reshape(-1, 1)
    X_test = test_df[features].values
    y_test = test_df[target].values.reshape(-1, 1)
    print(f"X_train shape: {X_train.shape}, X_test shape: {X_test.shape}, y_train shape: {y_train.shape}, y_test shape: {y_test.shape}")

    # Initialize and train TabNet
    tabnet = TabNetRegressor(verbose=0, seed=42)
    start_time = time.time()
    tabnet.fit(
        X_train, y_train,
        eval_set=[(X_test, y_test)],
        max_epochs=1000,
        patience=10,
        batch_size=256,
        virtual_batch_size=128
    )
    end_time = time.time()
    print(f"Training completed in {end_time - start_time:.2f} seconds for Fold {fold}")

    # Extract best validation MSE
    best_mse = min(tabnet.history['val_0_mse'])
    mse_scores.append(best_mse)

    # Predict and calculate metrics
    y_pred = tabnet.predict(X_test)
    y_test_flat = y_test.flatten()
    y_pred_flat = y_pred.flatten()
    
    # R²
    r2 = 1 - np.sum((y_test_flat - y_pred_flat) ** 2) / np.sum((y_test_flat - np.mean(y_test_flat)) ** 2)
    r2_scores.append(r2)
    
    # MAPE on log scale
    mape_log = np.mean(np.abs((y_test_flat - y_pred_flat) / (y_test_flat + 1e-6))) * 100
    mape_scores_log.append(mape_log)
    
    # MAPE on original scale
    y_test_orig = np.expm1(y_test_flat)
    y_pred_orig = np.expm1(y_pred_flat)
    mape_orig = np.mean(np.abs((y_test_orig - y_pred_orig) / (y_test_orig + 1e-6))) * 100
    mape_scores_orig.append(mape_orig)
    
    print(f"Fold {fold} - R²: {r2:.4f}, MSE: {best_mse:.4f}, MAPE (log): {mape_log:.2f}%, MAPE (original): {mape_orig:.2f}%")

    # Per-fold visualizations
    plt.figure(figsize=(12, 6))
    plt.plot(y_test_orig, label='Actual', color='blue')
    plt.plot(y_pred_orig, label='Predicted', color='orange')
    plt.title(f'Fold {fold}: Actual vs. Predicted sum_requested_prbs (Original Scale)')
    plt.xlabel('Time Steps')
    plt.ylabel('sum_requested_prbs')
    plt.legend()
    plt.grid(True)
    plt.show()

    plt.figure(figsize=(12, 6))
    plt.scatter(y_test_orig, y_test_orig - y_pred_orig, alpha=0.5)
    plt.axhline(0, color='red', linestyle='--')
    plt.xlabel('Actual sum_requested_prbs')
    plt.ylabel('Residuals')
    plt.title(f'Fold {fold}: Residuals')
    plt.grid(True)
    plt.show()

# Compute cross-validated metrics
mean_r2 = np.mean(r2_scores)
std_r2 = np.std(r2_scores)
mean_mse = np.mean(mse_scores)
std_mse = np.std(mse_scores)
mean_mape_log = np.mean(mape_scores_log)
std_mape_log = np.std(mape_scores_log)
mean_mape_orig = np.mean(mape_scores_orig)
std_mape_orig = np.std(mape_scores_orig)
print(f"\nCross-Validated Metrics (5-fold):")
print(f"Mean R²: {mean_r2:.4f} ± {std_r2:.4f}")
print(f"Mean MSE: {mean_mse:.4f} ± {std_mse:.4f}")
print(f"Mean MAPE (log scale): {mean_mape_log:.2f}% ± {std_mape_log:.2f}%")
print(f"Mean MAPE (original scale): {mean_mape_orig:.2f}% ± {std_mape_orig:.2f}%")

# Final test set evaluation
X_train_final = df.iloc[:int(0.9 * len(df))][features].values
X_test_final = df.iloc[int(0.9 * len(df)):][features].values
y_train_final = df.iloc[:int(0.9 * len(df))][target].values.reshape(-1, 1)
y_test_final = df.iloc[int(0.9 * len(df)):][target].values.reshape(-1, 1)
tabnet.fit(
    X_train_final, y_train_final,
    eval_set=[(X_test_final, y_test_final)],
    max_epochs=100,
    patience=10,
    batch_size=1024,
    virtual_batch_size=128,
    num_workers=0,
    drop_last=False
)
y_pred_final = tabnet.predict(X_test_final)

# Inverse transform for visualization
y_test_orig = np.expm1(y_test_final.flatten())
y_pred_orig = np.expm1(y_pred_final.flatten())

# Final test set visualizations
plt.figure(figsize=(12, 6))
plt.plot(y_test_orig, label='Actual', color='blue')
plt.plot(y_pred_orig, label='Predicted', color='orange')
plt.title('TabNet Prediction of sum_requested_prbs (Final Test Set)')
plt.xlabel('Time Steps')
plt.ylabel('sum_requested_prbs')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
plt.scatter(y_test_orig, y_test_orig - y_pred_orig, alpha=0.5)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Actual sum_requested_prbs')
plt.ylabel('Residuals')
plt.title('TabNet Residuals (Final Test Set)')
plt.grid(True)
plt.show()
```
